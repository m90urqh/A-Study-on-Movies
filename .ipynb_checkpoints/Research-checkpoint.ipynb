# Visualization
* To better understand the dataset, we visualize the distribution of movie budgets and ratings across the years. This graphical representation highlights how production costs and audience reception have evolved over time, making it easier to identify trends, fluctuations, and potential correlations between the two variables.

```python
import matplotlib.pyplot as plt
fig,ax = plt.subplots(1,3,figsize=(16,4))
avg_budget_per_year = data_file.groupby("Year")["Budget"].mean()
avg_rat_per_year = data_file.groupby("Year")["RottenTomatoes"].mean()

ax[0].plot(avg_budget_per_year.index, avg_budget_per_year.values,marker ="o",color = "red")
ax[0].plot(avg_rat_per_year.index,avg_rat_per_year.values,marker="s",color = "blue")
ax[0].set_xlabel("Year")
ax[0].set_ylabel("Budget: red - Rating: blue")
ax[0].set_title("movies average budgets and ratings through years")
ax[1].plot(avg_budget_per_year.index,avg_budget_per_year.values,marker="o",color = "red")
ax[1].set_xlabel("Year")
ax[1].set_ylabel("Budget")
ax[1].set_title("movies average budgets through years")
ax[2].plot(avg_rat_per_year.index,avg_rat_per_year.values,marker="s",color = "blue")
ax[2].set_xlabel("Year")
ax[2].set_ylabel("Rating")
ax[2].set_title("movies average ratings through years")
plt.show()
```

From the previous figures, we observe that movie budgets were significantly high in the early 2000s, dropped sharply around 2008, and then reached their peak in 2012. Interestingly, the rise in budgets appears to coincide with an increase in average ratings, suggesting that higher spending was often associated with better‑received films.

* We visualize the distribution of movie ratings across different genres for the year 2010. This snapshot highlights how audience reception varied among genres during that specific year, offering insights into which categories achieved higher ratings and which faced more critical challenges.

```python

movie_2010 = data_file[data_file['Year']==2010]
avg_rat_per_genere = movie_2010.groupby("Genre")["RottenTomatoes"].mean()

avg_rat_per_genere.plot(kind="bar",color = "black")
plt.xlabel("Genere")
plt.ylabel("Rating")
plt.title("movies average ratings through genere")
plt.show
```

From the 2010 genre ratings figure, we can see that Biography and Mystery films achieved the highest average ratings compared to other genres. Following them, Adventure and Animation movies also performed strongly, though with slightly lower averages.

* After examining ratings by genre for the year 2010, we now extend the analysis to cover the entire dataset. This visualization illustrates how average ratings evolve across genres over multiple years, revealing long‑term trends, fluctuations, and shifts in audience preferences. By comparing genres over time, we gain a broader perspective on which categories maintain consistent reception and which experience notable changes.

```python

avg_rat_per_year_genre = data_file.groupby(["Year","Genre"])["RottenTomatoes"].mean().unstack()
avg_rat_per_year_genre.plot(kind="line", marker="o")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.title("Average Ratings by Genre Over Time")
plt.legend(title="Genre", bbox_to_anchor=(1.05, 1), loc="upper left")
plt.show()
```

From this figure, we observe how movie genre ratings shifted over the years. By 2013, audience interest in romance films had noticeably declined, while ratings for horror movies showed the opposite trend, reflecting a growing appeal in that genre.

### Finally, we extend the analysis to include predictions of average ratings for upcoming movies. This forecast is based on historical trends observed in the dataset and provides an estimate of how audience reception may evolve in future releases. While the prediction offers a mathematical perspective on potential rating outcomes, it should be interpreted with caution, as real‑world factors such as genre shifts, production quality, and audience preferences can significantly influence actual results.

```python
avg_rat_th_year = data_file.groupby("Year")["RottenTomatoes"].mean()
years = avg_rat_th_year.index.values.reshape(-1,1)
rating = avg_rat_th_year.values

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

poly = PolynomialFeatures(degree=2)
x_poly = poly.fit_transform(years)
model = LinearRegression()
model.fit(x_poly,rating)
```

```python
import numpy as np
fu_years = np.array([2014,2015,2016,2017]).reshape(-1,1)
fu_poly = poly.fit_transform(fu_years)
pred = model.predict(fu_poly)
print("the predicted values of the avg of the ratings of the movies in the future is :",pred)
```

#### We visualize the historical average ratings across the years alongside the predicted values. This combined view allows us to compare past audience reception with the forecasted trend, highlighting how the model projects ratings to evolve. By placing actual averages and predicted averages together, the visualization provides a clearer perspective on the reliability and direction of the prediction.

```python
plt.plot(avg_rat_per_year.index,avg_rat_per_year.values,marker = "o",color = "blue",label = "old avg ratings")
plt.plot(fu_years,pred,marker="s",color= "red",label = "predicited avg ratings")
plt.title("avg ratings through years")
plt.xlabel("Year")
plt.ylabel("Rating")
plt.legend()
```

This prediction is purely mathematical: based on past years, the model suggests that the average movie rating will continue to rise, reaching around 80 by 2017. It reflects a simple linear regression forecast, which assumes a steady upward trend.
